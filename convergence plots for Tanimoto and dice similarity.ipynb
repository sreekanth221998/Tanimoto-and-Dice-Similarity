```python
import pandas as pd
import matplotlib.pyplot as plt

# load the data from a csv file
df = pd.read_csv(r"E:\pytesseract\POLYMER DATA FILES\PolyGNN\Tanimoto and Dice csvs\Tanimoto_similarity1.csv")

# create a list of the polymer names
# polymer_names = df['Name'].unique() # all names
polymer_names = df['Name'].unique()[:3] # Only keep the first 3 names
# create a figure and axis for the plot
fig, ax = plt.subplots()

# loop through each polymer
for Name in polymer_names:
    # select the data for the current polymer
    data = df[df['Name'] == Name]
    # create a list of the repeating units
    x = list(range(1,51))
    # create a list of the Tanimoto similarity coefficients
    y = data.iloc[0,1:].tolist()
    # plot the data
    ax.plot(x, y, label=Name)

# add axis labels and a legend
ax.set_xlabel('Number of Repeating Units')
ax.set_ylabel('Tanimoto Similarity Coefficient')
ax.legend()
plt.xlim(2,50)

# add vertical line at 16th point on x-axis
ax.axvline(x=16, color='red', linestyle='--')

# plt.savefig(f'{Name}.png')
plt.savefig('Tanimoto_similarity.png')

# show the plot
plt.show()
```
